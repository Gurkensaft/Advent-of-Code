```python
import re
from collections import defaultdict
engine = open('engine.txt').read().splitlines()
engine = [engine_line.split('\n')[0] for engine_line in engine]

front_engine = ['.'*(len(engine[0])+2)]
back_engine = front_engine.copy()
engine = front_engine + ['.'+line+'.' for line in engine] + back_engine
```

```python
def extract_number(line):
    return re.findall(r'\d+', line)[0]
```

```python
def checkForSymbol(list):
    for i in list:
        if i not in ['.', '1', '2', '3', '4', '5', '6', '7', '8', '9', '0']:
            return True
    return False
```

```python
def checkIfSurrounded(engine, roadnumber, position):
    #first road
    if roadnumber == 0:
        #first position
        if position == 0:
            return checkForSymbol([engine[roadnumber][position+1],engine[roadnumber+1][position],engine[roadnumber+1][position+1]])
        #last position
        if position == len(engine[roadnumber])-1:
            return checkForSymbol([engine[roadnumber][position-1],engine[roadnumber+1][position],engine[roadnumber+1][position-1]])
        #between position
        return checkForSymbol([engine[roadnumber][position-1],engine[roadnumber][position+1],engine[roadnumber+1][position],engine[roadnumber+1][position-1],engine[roadnumber+1][position+1]])
    #last road
    if roadnumber == len(engine)-1:
        #first position
        if position == 0:
            return checkForSymbol([engine[roadnumber][position+1],engine[roadnumber-1][position],engine[roadnumber-1][position+1]])
        #last position
        if position == len(engine[roadnumber])-1:
            return checkForSymbol([engine[roadnumber][position-1],engine[roadnumber-1][position],engine[roadnumber-1][position-1]])
        #between position
        return checkForSymbol([engine[roadnumber][position-1],engine[roadnumber][position+1],engine[roadnumber-1][position],engine[roadnumber-1][position-1],engine[roadnumber-1][position+1]])
    #between road
    #first position
    if position == 0:
        return checkForSymbol([engine[roadnumber][position+1],engine[roadnumber-1][position],engine[roadnumber-1][position+1],engine[roadnumber+1][position],engine[roadnumber+1][position+1]])
    #last position
    if position == len(engine[roadnumber])-1:
        return checkForSymbol([engine[roadnumber][position-1],engine[roadnumber-1][position],engine[roadnumber-1][position-1],engine[roadnumber+1][position],engine[roadnumber+1][position-1]])
    #between position
    return checkForSymbol([engine[roadnumber][position-1],engine[roadnumber][position+1],engine[roadnumber-1][position],engine[roadnumber-1][position-1],engine[roadnumber-1][position+1],engine[roadnumber+1][position],engine[roadnumber+1][position-1],engine[roadnumber+1][position+1]])
```

```python
numbers = []
for roadnumber,road in enumerate(engine):
    number_touched = False
    surrounded = False
    for position,char in enumerate(road):
        if char == '.':
            if number_touched and surrounded:
                numbers.append(extract_number(road[start_position:position]))
                number_touched = False
                surrounded = False
            if number_touched and not surrounded:
                number_touched = False
        elif char.isnumeric():
            if not number_touched:
                start_position = position
                number_touched = True
            if number_touched and not surrounded:
                surrounded = checkIfSurrounded(engine,roadnumber,position)
        else:
            if number_touched:
                numbers.append(extract_number(road[start_position:position]))
                number_touched = False
                surrounded = False
            
```

```python

surround = defaultdict(list)
for i, road in enumerate(engine):
    for m in re.finditer(r"\d+", road):
        pos = [(i, m.start() - 1), (i, m.end())]
        #checking surroundings of current position
        pos += [(i - 1, j) for j in range(m.start() - 1, m.end() + 1)] 
        pos += [(i + 1, j) for j in range(m.start() - 1, m.end() + 1)]

        for a, b in pos:
            if 0 <= a < len(engine) and 0 <= b < len(engine[a]) and engine[a][b] != ".":
                surround[a, b].append(m.group())
res = sum(int(x[0]) * int(x[1]) for x in surround.values() if len(x) == 2)
```
