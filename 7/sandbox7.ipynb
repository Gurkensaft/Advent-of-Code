```python
import re
import pandas as pd
import numpy as np
input = open("input.txt", "r").read().strip().split("\n")
hands_and_bids = [line.split(" ") for line in input]
```

```python
df = pd.DataFrame(hands_and_bids, columns=["hand", "bid"])
```

```python
ranks = {'A':14,
        'K':13, 
        'Q':12, 
        'J':1,
        'T':10,
        '9':9,
        '8':8,
        '7':7,
        '6':6,
        '5':5,
        '4':4,
        '3':3,
        '2':2}
```

```python
def get_possible_ranks(combs, jokers):
    all_ranks = []
    combs.append(0)
    og_combs = combs.copy()
    for i in range(len(combs)):
        combs = og_combs.copy()
        combs[i] = combs[i] + jokers

        if 5 in combs:
            all_ranks.append(7)
        elif 4 in combs:
            all_ranks.append(6)
        elif 3 in combs and 2 in combs:
            all_ranks.append(5)
        elif 3 in combs:
            all_ranks.append(4)
        elif combs.count(2) == 2:
            all_ranks.append(3)
        elif 2 in combs:
            all_ranks.append(2)
        else:
            all_ranks.append(1)
    return all_ranks
```

```python
def get_hand_rank(hand):
    combs=[]
    jokers = 0
    for key in ranks.keys():
        occurences = hand.count(key)
        if occurences > 0 and key != 'J':
            combs.append(occurences)
        if key == 'J':
            jokers = occurences
    return max(get_possible_ranks(combs, jokers))

    
```

```python
hand_ranks = []
for hb in df.itertuples():
    hand_ranks.append(get_hand_rank(hb.hand))
print(hb)
```

```python
df['hand_rank'] = hand_ranks
df.sort_values(by=['hand_rank'], ascending=False)
```

```python
def get_finger_value(finger, hand):
    return ranks[hand[finger]]

first_finger = [get_finger_value(0, hb.hand) for hb in df.itertuples()]
second_finger = [get_finger_value(1, hb.hand) for hb in df.itertuples()]
third_finger = [get_finger_value(2, hb.hand) for hb in df.itertuples()]
fourth_finger = [get_finger_value(3, hb.hand) for hb in df.itertuples()]
fifth_finger = [get_finger_value(4, hb.hand) for hb in df.itertuples()]
df['first_finger'] = first_finger
df['second_finger'] = second_finger
df['third_finger'] = third_finger
df['fourth_finger'] = fourth_finger
df['fifth_finger'] = fifth_finger
```

```python
df = df.sort_values(by=['hand_rank', 'first_finger', 'second_finger', 'third_finger', 'fourth_finger', 'fifth_finger'], ascending=False).reset_index(drop=True)
```
